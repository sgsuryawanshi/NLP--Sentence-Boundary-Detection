```python
import pandas as pd
import re
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
#Extracting the data from the train & test data
with open("SBD.train") as file1:
    d_train = file1.read().splitlines()
with open("SBD.test") as file2:
    d_test = file2.read().splitlines()
d_test
```

```python
# Inserting the data into the organised list format
dd_train = []
for n in range(0,len(d_train)):
    dd_train+=[re.findall(r'\S+',d_train[n])]
dd_test = []
for n in range(0,len(d_test)):
    dd_test+=[re.findall(r'\S+',d_test[n])]
dd_test
```

```python
#Extracting the values of attributes to detect the output
regex=re.compile(r'\w+\.')
L = []
for i in range(0,len(dd_train)):
    if dd_train[i][1].endswith('.'):
        if (i != (len(dd_train)-1)):
            L.append([dd_train[i][2] , dd_train[i][1][:-1] , dd_train[i+1][1] , int(len(dd_train[i][1][:-1])<3) , int(dd_train[i][1][0].isupper()) , int(dd_train[i+1][1][0].isupper()) , int(dd_train[i+1][1][0] == '"') , int(isinstance(dd_train[i+1][1][0], int)) , int(len(dd_train[i+1][1])>2) ])
        else:
            L.append([dd_train[i][2] , dd_train[i][1][:-1] , dd_train[i][1][:-1] , int(len(dd_train[i][1][:-1])<3) , int(dd_train[i][1][0].isupper()) , int(dd_train[i][1][0].isupper()) , int(dd_train[i][1][0] == '"') , int(isinstance(dd_train[i][1][0], int)) , int(len(dd_train[i][1])>2) ])

L_test = []
for i in range(0,len(dd_test)):
    if dd_test[i][1].endswith('.'):
        if (i != (len(dd_test)-1)):
            L_test.append([dd_test[i][2] , dd_test[i][1][:-1] , dd_test[i+1][1] , int(len(dd_test[i][1][:-1])<3) , int(dd_test[i][1][0].isupper()) , int(dd_test[i+1][1][0].isupper()) , int(dd_test[i+1][1][0] == '"') , int(isinstance(dd_test[i+1][1][0], int)) , int(len(dd_test[i+1][1])>2) ])
        else:
            L_test.append([dd_test[i][2] , dd_test[i][1][:-1] , dd_test[i][1][:-1] , int(len(dd_test[i][1][:-1])<3) , int(dd_test[i][1][0].isupper()) , int(dd_test[i][1][0].isupper()) , int(dd_test[i][1][0] == '"') , int(isinstance(dd_test[i][1][0], int)) , int(len(dd_test[i][1])>2)])
```

```python
for i in L_test:
    print(i)
```

```python
#Creating the Dataframes for the test data & train data
train = pd.DataFrame(L, columns = ['EOS/NEOS', 'Left','Right','L < 3', 'L_capital','R_capital','R = "' , 'R: Number' , 'R > 2'])
test = pd.DataFrame(L_test, columns = ['EOS/NEOS', 'Left','Right','L < 3', 'L_capital','R_capital','R = "' , 'R: Number' , 'R > 2'])
```

```python
train
```

```python
# Inorder to make the DF compatible with Decison Trees, we restore the left word of the period & right word of the period with index
train["Left"]=train.index
train["Right"]=train.index
test["Left"]=test.index
test["Right"]=test.index

train_data=train.replace({"EOS":1,"NEOS":0})
test_data=test.replace({"EOS":1,"NEOS":0})
```

```python
# Initial Features - Dataset
y=train_data[["EOS/NEOS"]]
x=train_data[["Left","Right","L < 3","L_capital","R_capital"]]
test_data_F=train_data[["Left","Right","L < 3","L_capital","R_capital"]]
test_data_F1=train_data[["EOS/NEOS"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(
        x, y, test_size=0.5,stratify=y, random_state=1
    )
```

```python
model = DecisionTreeClassifier(max_depth=2,random_state=1)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
testprediction=model.predict(test_data_F)
test_accuracy=accuracy_score(test_data_F1,testprediction)
print("Accuracy for core features:",test_accuracy*100)
```

```python
# The training model for all the 8 attributes

print("Training model used for 8 features")
y_new=train_data[["EOS/NEOS"]]
x_new=train_data[['Left','Right','L < 3','L_capital','R_capital','R = "','R: Number','R > 2']]
test_data_F_new=train_data[['Left','Right','L < 3','L_capital','R_capital','R = "','R: Number','R > 2']]
test_data_F1_new=train_data[["EOS/NEOS"]]

x_train_new, x_test_new, y_train_new , y_test_new = train_test_split(x_new, y_new , test_size = 0.5, stratify= y_new, random_state = 1)
model_new = DecisionTreeClassifier(max_depth=2,random_state=1)
```

```python
model_new.fit(x_train_new,y_train_new)
```

```python
y_pred_new=model_new.predict(x_test_new)
acc=accuracy_score(y_test_new,y_pred_new)
print("The training accuracy for all the 8 attributes is:",round((acc*100),2),"%")
```

```python
# Predicting the accuracy for the Test Data
test_pred_new=model_new.predict(test_data_F_new)
testaccuracy_new=accuracy_score(test_data_F1_new,test_pred_new)
print("The accuracy for initial Features is:",round(testaccuracy_new*100,2),"%")
```

```python
#The Training model for newly implemented attributes
y_features=train_data[["EOS/NEOS"]]
x_features=train_data[['R = "','R: Number','R > 2']]
test_data_F_features=train_data[['R = "','R: Number','R > 2']]
test_data_F1_features=train_data[["EOS/NEOS"]]

x_train_features, x_test_features, y_train_features, y_test_features = train_test_split(
        x_features, y_features, test_size=0.5,stratify=y_features, random_state=1
    )
model_features = DecisionTreeClassifier(max_depth=2,random_state=1)
```

```python
model_features.fit(x_train_features,y_train_features)

y_pred_features=model_features.predict(x_test_features)
acc_features=accuracy_score(y_test_features,y_pred_features)
print(" The Training Accuracy for 3 new features are:",round((acc_features*100),2),"%")
```

```python
#Predicting the acccuracy of Test data
test_pred_features=model_features.predict(test_data_F_features)
testaccuracy_3=accuracy_score(test_data_F1_features,test_pred_features)
print("The Accuracy of 3  newly added features:",round(testaccuracy_3*100,2),"%")
```
